```python
#Analisis Data Ekonomi
def f_outliers(x):
    Q3 = np.percentile(x,75)
    Q1 = np.percentile(x,25)
    IQR = Q3-Q1
    df_final = x[(x > (Q3+(1.5*IQR)))]
    return df_final
def remove_outliers(x):
    Q3 = np.percentile(x,75)
    Q1 = np.percentile(x,25)
    IQR = Q3-Q1
    df_ro = x[~(x>(Q3+(IQR*1.5)))]
    return df_ro
```

```python
print(f_outliers(df_1['Pendapatan']))
print(f_outliers(df_1["GDPperkapita"]))
```

```python
df_2 = remove_outliers(df_1[["GDPperkapita","Pendapatan"]])
df_2.dropna(axis=0,inplace=True)
df_2
```

```python
sns.boxplot(data=df_2,x="GDPperkapita");
```

```python
sns.boxplot(data=df_2,x="Pendapatan");
```

Karena masih terdapat data outliers pada data GDP Perkapita yang melebihi angka 29.000 (jauh di atas rata-rata) dan pada data pendapatan yang melebihi angka 40.000. Maka akan dilakukan penghilangan outliers kembali untuk memperoleh label kategori Ekonomi suatu negara secara lebih baik lagi.

```python
df_3 = remove_outliers(df_2[["GDPperkapita","Pendapatan"]])
df_3.dropna(axis=0,inplace=True)
df_3.info()
```

```python
sns.boxplot(data=df_3,x="GDPperkapita");
```

```python
sns.boxplot(data=df_3,x="Pendapatan");
```

Agar data tidak hilang semua, maka dibatasi proses penghilangan outliersnya sampai 2 kali saja.

```python
#Clustering data
#Merapikan data
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
sc = StandardScaler()
df_3_std = sc.fit_transform(df_3)
#Elbor Method:
wcss = []
for i in range (1,11):
    kmeans = KMeans(n_clusters=i, init="k-means++", random_state = 42)
    kmeans.fit(df_3_std)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow method')
plt.xlabel('n_clusters')
plt.ylabel('wcss')
plt.show
```

```python
#Penggabungan Data
kmeans_1 = KMeans(n_clusters=3, init="k-means++", random_state=42).fit(df_3_std)
labels_1 = kmeans_1.labels_
new_df_3 = pd.DataFrame(data=df_3_std, columns=['GDPperkapita', 'Pendapatan'])
new_df_3['Tingkat_Ekonomi'] = labels_1
new_df_3
```

```python
#Visualisasi
plt.figure(figsize=(17,10))
plt.scatter(new_df_3["Pendapatan"][new_df_3["Tingkat_Ekonomi"]==0],
            new_df_3["GDPperkapita"][new_df_3["Tingkat_Ekonomi"]==0],
           color = "r", s = 100, label = "Negara Kaya")
plt.scatter(new_df_3["Pendapatan"][new_df_3["Tingkat_Ekonomi"]==1], 
            new_df_3["GDPperkapita"][new_df_3["Tingkat_Ekonomi"]==1],
           color = "b", s = 100, label = "Negara Miskin")
plt.scatter(new_df_3["Pendapatan"][new_df_3["Tingkat_Ekonomi"]==2], 
            new_df_3["GDPperkapita"][new_df_3["Tingkat_Ekonomi"]==2],
           color = "g", s = 100, label = "Negara Berkembang")
plt.scatter(kmeans_1.cluster_centers_[:,0], kmeans_1.cluster_centers_[:,0],c="k", s=300)
plt.xlabel("Pendapatan")
plt.ylabel("GDP Perkapita")
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeansa = KMeans(n_clusters = 3, random_state=42)
kmeansa.fit(sc.inverse_transform(df_3_std))
labels_1_1 = kmeansa.labels_
labels_1_1
```

```python
df_3["Tingkat_Ekonomi"] = labels_1_1
```

```python
#Visualisasi_Normal
plt.figure(figsize=(17,10))
plt.scatter(df_3["Pendapatan"][df_3["Tingkat_Ekonomi"]==0],
            df_3["GDPperkapita"][df_3["Tingkat_Ekonomi"]==0],
           color = "b", s = 100, label = "Negara Kaya")
plt.scatter(df_3["Pendapatan"][df_3["Tingkat_Ekonomi"]==1], 
            df_3["GDPperkapita"][df_3["Tingkat_Ekonomi"]==1],
           color = "r", s = 100, label = "Negara Miskin")
plt.scatter(df_3["Pendapatan"][df_3["Tingkat_Ekonomi"]==2], 
            df_3["GDPperkapita"][df_3["Tingkat_Ekonomi"]==2],
           color = "g", s = 100, label = "Negara Berkembang")
plt.scatter(kmeansa.cluster_centers_[:,0], kmeansa.cluster_centers_[:,1],c="k", s=300)
plt.xlabel("Pendapatan")
plt.ylabel("GDP Perkapita")
plt.legend()
plt.show()
```
